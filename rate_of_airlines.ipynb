```python
airlines_df = pd.read_csv('airlines.csv')
airlines_df
```

```python
df_new = pd.merge(df,airlines_df, left_on = 'AIRLINE', right_on = 'IATA_CODE')
df_new
```

```python
df_new.rename(columns={'AIRLINE_x': 'IATA_CODES', 'AIRLINE_y': 'AIRLINES'}, inplace=True)
df_new.drop(columns=['IATA_CODE'], inplace=True)
```

```python
df_new
```

```python
climate_conditions = [
    (df_new['MONTH'] < 3 ) | (df_new['MONTH'] == 12),
    (df_new['MONTH'] >= 3) & (df_new['MONTH'] < 6),
    (df_new['MONTH'] >= 6) & (df_new['MONTH'] < 9),
    (df_new['MONTH'] >= 9) & (df_new['MONTH'] < 12)
    ]
climate_catagories = ['Winter', 'Spring', 'Summer', 'Autumn']
df_new['CLIMATE'] = np.select(climate_conditions, climate_catagories)
df_new.sample(10)
```

```python
flight_numbers = df_new.AIRLINES.value_counts()
flight_numbers
```

```python
df_new.groupby("AIRLINES").CANCELLATION_REASON.value_counts()
```

```python
cancels = df_new[df_new.CANCELLATION_REASON == 'A'].groupby(['AIRLINES']).CANCELLED.count()
cancels
```

```python
cancel_rates = cancels/flight_numbers
cancel_rates.sort_values()
```

```python
delays = df_new[df_new.AIRLINE_DELAY + df_new.LATE_AIRCRAFT_DELAY  > 0].groupby(['AIRLINES']).AIRLINE_DELAY.count()
delays
```

```python
delay_rates = delays/flight_numbers
delay_rates.sort_values()
```

```python
toplam_rates = cancel_rates + delay_rates
toplam_rates.sort_values()
```

```python
df_toplam_rates = pd.DataFrame(toplam_rates)
df_toplam_rates = df_toplam_rates.rename(columns={0: 'RATES'})
df_toplam_rates = df_toplam_rates.reset_index()
df_toplam_rates = df_toplam_rates.rename(columns={'index': 'AIRLINES'})
df_toplam_rates
```

```python
general_rates_table = sns.barplot(x = "RATES", y = "AIRLINES", data= df_toplam_rates);
general_rates_table.set(xlim=(0.060, 0.210))
general_rates_table.set_xlabel('Airlines', fontsize=20)
general_rates_table.set_ylabel('Rates', fontsize=20)
plt.title('Percentage of Cancellations and Delays', fontsize=18)
general_rates_table;
```

```python
cancels_climates = df_new[df_new.CANCELLATION_REASON == 'A'].groupby(['CLIMATE','AIRLINES']).CANCELLED.count()
cancels_climates
```

```python
flights_climate = df_new.groupby(['CLIMATE', 'AIRLINES'])['YEAR'].count()
df_flights_climate = pd.DataFrame(flights_climate)
df_flights_climate
df_flights_climate = df_flights_climate.rename(columns={'YEAR': 'FLIGHT_NUMBER'})
df_flights_climate
```

```python
cancels_climate_rates = cancels_climates/flights_climate
df_cancels_climate_rates = pd.DataFrame(cancels_climate_rates)
df_cancels_climate_rates = df_cancels_climate_rates.rename(columns={0: 'RATES'})
df_cancels_climate_rates
```

```python
delays_climate = df_new[df_new.AIRLINE_DELAY + df_new.LATE_AIRCRAFT_DELAY  >= 0].groupby(['CLIMATE','AIRLINES']).AIRLINE_DELAY.count()
df_delays_climate = pd.DataFrame(delays_climate)
df_delays_climate
```

```python
delays_climate_rates = delays_climate/flights_climate
df_delays_climate_rates = pd.DataFrame(delays_climate_rates)
df_delays_climate_rates
df_delays_climate_rates = df_delays_climate_rates.rename(columns={0: 'RATES'})
df_delays_climate_rates
```

```python
total_climate_rates=cancels_climate_rates+delays_climate_rates
df_total_climate_rates = pd.DataFrame(total_climate_rates)
df_total_climate_rates
df_total_climate_rates = df_total_climate_rates.rename(columns={0: 'RATES'})
df_total_climate_rates = df_total_climate_rates.reset_index()
df_total_climate_rates
```

```python
general_climate_rates_table = sns.barplot(x="AIRLINES", y="RATES", hue="CLIMATE", 
                                         data=df_total_climate_rates, palette=['Green','Blue', 'Red', 'Yellow'])
general_climate_rates_table.set_xticklabels(general_climate_rates_table.get_xticklabels(), rotation=90)
general_climate_rates_table.set(ylim=(0.060, 0.400))
general_climate_rates_table.set_xlabel('Airlines', fontsize=20)
general_climate_rates_table.set_ylabel('Delay&Canc. Rates', fontsize=20)
general_climate_rates_table;
```

```python
df_Autumn_rates = df_total_climate_rates[df_total_climate_rates.CLIMATE == 'Autumn']
df_Autumn_rates
```

```python
import matplotlib.ticker as mtick
Autumn_rates_table = sns.barplot(x="AIRLINES", y="RATES",
                                         data=df_Autumn_rates)
Autumn_rates_table.set_xticklabels(Autumn_rates_table.get_xticklabels(), rotation=90)
Autumn_rates_table.set(ylim=(0.050, 0.250))
Autumn_rates_table.set_xlabel('Airlines', fontsize=20)
Autumn_rates_table.set_ylabel('Delay&Canc. Rates', fontsize=20)
plt.title('Autumn', fontsize=25, color='green')
plt.figure(figsize=(12, 7))
Autumn_rates_table.yaxis.set_major_formatter(mtick.PercentFormatter(1,0))
for p in Autumn_rates_table.patches:
    Autumn_rates_table.annotate(format(p.get_height(), '1.2%'), 
                   (p.get_x() + p.get_width() / 2., p.get_height()), 
                   ha = 'center', va = 'top',
                   size=9,
                   xytext = (0, 10), 
                   textcoords = 'offset points')
Autumn_rates_table;
```
